```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Day_9_banking_data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
grouped = df.groupby('Account_Type')
transaction_sum = grouped['Transaction_Amount'].sum()
account_balance_avg = grouped['Account_Balance'].mean()
print("Total Transaction Amount by Account Type:\n", transaction_sum)
print("\nAverage Account Balance by Account Type:\n", account_balance_avg)
```

```python
grouped = df.groupby('Branch')
total_transactions = grouped['Transaction_Amount'].count()
average_transaction_amount = grouped['Transaction_Amount'].mean()
print("Total Number of Transactions per Branch:\n", total_transactions)
print("\nAverage Transaction Amount per Branch:\n", average_transaction_amount)
```
